```python
def run_quiz(questions):
    score = 0

    print("üß† Welcome to the Quiz Game!\n")

    for i, q in enumerate(questions, 1):
        print(f"Q{i}: {q['question']}")
        for option in q['options']:
            print(option)
        
        answer = input("Your answer (A/B/C/D): ").strip().upper()
        
        if answer == q['answer']:
            print("‚úÖ Correct!\n")
            score += 1
        else:
            print(f"‚ùå Wrong! The correct answer is {q['answer']}.\n")
    
    print(f"üéâ Quiz Completed! You scored {score} out of {len(questions)}.")

# List of questions
quiz_questions = [
    {
        "question": "What is the capital of France?",
        "options": ["A. Berlin", "B. Madrid", "C. Paris", "D. Rome"],
        "answer": "C"
    },
    {
        "question": "Which planet is known as the Red Planet?",
        "options": ["A. Earth", "B. Mars", "C. Venus", "D. Jupiter"],
        "answer": "B"
    },
    {
        "question": "Who wrote the play 'Romeo and Juliet'?",
        "options": ["A. Charles Dickens", "B. William Shakespeare", "C. Mark Twain", "D. Jane Austen"],
        "answer": "B"
    },
    {
        "question": "what is the capital of India?",
        "options": ["A. New Delhi", "B. Paris", "C. Tokyo", "D. America"],
        "answer": "A"
    },
    {
        "question": "What is the natural satellite of earth?",
        "options": ["A. sun", "B. moon", "C. stars", "D. artificial satellite"],
        "answer": "B"
    }
]

# Start the quiz
run_quiz(quiz_questions)
```
